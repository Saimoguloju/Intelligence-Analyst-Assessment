```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Create a geometry column with Point objects
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]

# Convert the DataFrame to a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Set the coordinate reference system (CRS) to WGS84 (common for GPS data)
gdf.set_crs(epsg=4326, inplace=True)
```

```python
import folium
from folium.plugins import HeatMap

# Initialize a folium map centered on a key point
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

# Create a list of locations for the heatmap
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]

# Add a heatmap layer to the map
HeatMap(heat_data).add_to(m)

# Save the map to an HTML file
m.save('heatmap.html')
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Point
```

```python
# Load the dataset
df = pd.read_excel('Assingment_Dataset.xlsx')

# Convert the DataFrame to a GeoDataFrame
gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df.Longitude, df.Latitude),
    crs="EPSG:4326"  # WGS 84 coordinate reference system
)
```

```python
# Filter cross-border movements (adjust the filtering logic as needed)
cross_border_gdf = gdf[gdf['Movement Type'] == 'Cross-Border']
```

```python
# Set up the plot
plt.figure(figsize=(10, 10))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
ax = world.plot(figsize=(10, 10), color='white', edgecolor='black')

# Plot the cross-border movements as a heatmap
sns.kdeplot(
    x=cross_border_gdf.geometry.x,
    y=cross_border_gdf.geometry.y,
    cmap="Reds",
    fill=True,
    ax=ax
)

plt.title('Heatmap of Cross-Border Movements')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Check the first few rows of the filtered GeoDataFrame
print(cross_border_gdf.head())

# Check if the coordinates are present
print(cross_border_gdf[['geometry']].dropna().shape)
```

```python
# Extract coordinates
coords = np.array(list(zip(cross_border_gdf.geometry.x, cross_border_gdf.geometry.y)))

# Check the shape of coords
print(coords.shape)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Ensure the filtered data has coordinates
if not cross_border_gdf.empty:
    # Extract coordinates for clustering
    coords = np.array(list(zip(cross_border_gdf.geometry.x, cross_border_gdf.geometry.y)))
    
    # Check if the coords array is valid
    if coords.size > 0 and coords.shape[1] == 2:
        # Perform DBSCAN clustering
        db = DBSCAN(eps=0.5, min_samples=10).fit(coords)
        cross_border_gdf['Cluster'] = db.labels_
        
        # Plot the clusters
        plt.figure(figsize=(10, 10))
        world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
        ax = world.plot(figsize=(10, 10), color='white', edgecolor='black')
        cross_border_gdf.plot(ax=ax, column='Cluster', cmap='Set1', legend=True)
        plt.title('Hotspots of Cross-Border Movements')
        plt.show()
    else:
        print("No valid coordinates for clustering.")
else:
    print("No cross-border movements in the filtered data.")
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Ensure the CRS matches
if cross_border_gdf.crs != world.crs:
    cross_border_gdf = cross_border_gdf.to_crs(world.crs)

# Extract coordinates for clustering
coords = np.array(list(zip(cross_border_gdf.geometry.x, cross_border_gdf.geometry.y)))

# Check if the coords array is valid
if coords.size > 0 and coords.shape[1] == 2:
    # Perform DBSCAN clustering
    db = DBSCAN(eps=0.5, min_samples=10).fit(coords)
    cross_border_gdf['Cluster'] = db.labels_
    
    # Set up the plot
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Plot the world map
    world.plot(ax=ax, color='white', edgecolor='black')
    
    # Plot the clusters
    cross_border_gdf.plot(ax=ax, column='Cluster', cmap='Set1', legend=True, markersize=10)
    
    # Set aspect ratio to 'equal' to avoid distortion
    ax.set_aspect('equal')
    
    plt.title('Hotspots of Cross-Border Movements')
    plt.savefig('cross_border_hotspots.png')
    plt.show()
else:
    print("No valid coordinates for clustering.")
```
